```python
import pandas as pd

# Import population estimates as a data frame.
df = pd.read_csv("PopulationEstimates/Population Estimates 2010-2016-Table 1.csv", skiprows=2)
```

```python
# Preview the data frame.
df.head()
```

```python
# We only want the rows corresponding to states themselves, which all have multiples of 1000 as FIPS.
# Run a filter to check if the FIPS%1000 is 0 on preview.
df.head(100).loc[df['FIPS']%1000 == 0]
```

```python
# Run the same filter on the full data frame.
states = df.loc[df['FIPS']%1000 == 0]
```

```python
# This is pretty close, but we also included the full country row (United States), and Puerto Rico
states
```

```python
# It's crude, but let's just forcibly remove them.
states = states.loc[df['FIPS'] != 72000]
states = states.loc[df['FIPS'] != 0]
```
